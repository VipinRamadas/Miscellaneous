### 4. Correlation analysis

```python
import matplotlib.pyplot as plt         #importing matplotlib library and aliasing it to plt
```

```python
CRmatrix = df.corr()                    #Finding the correlation matrix of dataset and assigning it to variable CRmartix
plt.subplots(figsize=(10,4))            #Plotting the Correlation matrix
sns.heatmap(CRmatrix,annot=True,vmin=-2,vmax=1,linewidths=0.1,cmap='coolwarm')   #plotting a heatmap to show coorelation
```

```python
df['ph'].corr(df['Chloramines'])      #correlation factor between PH and Chloramines
```

```python
df['ph'].corr(df['Conductivity'])     #correlation factor between PH and Conductivity
```

#### inference on  correlation values 

##### 1.The linear association of ph to chloramines is very weak and negative. So when ph increases Chloramines decreases.


##### 2.The linear association of ph to Conduvtivity is very weak but positive.That is When ph increases Conductivity also increases.
