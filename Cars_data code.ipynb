## 1.Reading the dataset

## 3.Bar graph of male vs female buyers participated in the sales. 

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
x = ["Male","Female"]
y = [len(df[df['Buyer Gender']=="Male"]),len(df[df['Buyer Gender']=="Female"])]
plt.bar(x,y,width=0.8,color='green')
plt.title('Gender vs sales',color='red',fontsize=15)
plt.xlabel('Gender',color='blue',fontsize=13)
plt.ylabel('Sales',color='blue',fontsize=13)
plt.xticks()
plt.yticks()
```
