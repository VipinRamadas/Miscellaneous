### 3.Check whether all the zones generate the same amount of sales

Null Hypothesis : Same sales was generated by all the 4 zones  [H0: μ1=μ2=…=μp]

Alternate Hypothesis : Not the same sales was generated by all the 4 zones  [H1: All μ are not equal]

```python
from scipy.stats import f_oneway
```

```python
import scipy.stats as stats
```

```python
fvalue, pvalue = stats.f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'], df['Zone - D'])
print(fvalue, pvalue)
```

insights:

The p value obtained from ANOVA analysis is significant (p < 0.05), and therefore, we conclude that there are significant differences.i.e Alternate Hypothesis Stands,which means different sales was generated by sales.
